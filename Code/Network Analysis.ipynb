# Degree Centrality

```python
#Which technology tag are considered more central
nx.degree_centrality(G)
```

```python
#Highest degree centrality
degree_centrality = nx.degree_centrality(G)
max_centrality_value = max(degree_centrality.values())
max_centrality_nodes = [node for node, centrality in degree_centrality.items() if centrality == max_centrality_value]
print("Maximum degree centrality value:", max_centrality_value)
print("Nodes with maximum degree centrality:", max_centrality_nodes)
#jquery had the highest value for centrality
```

```python
#Lowest degree centrality
degree_centrality = nx.degree_centrality(G)
min_centrality_value = min(degree_centrality.values())
min_centrality_nodes = [node for node, centrality in degree_centrality.items() if centrality == min_centrality_value]
print("Minimum degree centrality value:", min_centrality_value)
print("Nodes with minimum degree centrality:", min_centrality_nodes)
```

```python
#Visualize centrality
centrality = nx.degree_centrality(G)
node_sizes = [1000 * v for v in centrality.values()]
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=node_sizes, node_color='plum', edge_color='gray')
plt.show()
```

# Social Network Extraction of jquery

```python
ego_node = 'jquery'
ego_net = nx.ego_graph(G, ego_node)

print("Nodes in Ego Network:", ego_net.nodes())
print("Edges in Ego Network:", ego_net.edges())
```

```python
#Visualize the ego network of the ego node 'jquery'
plt.figure(figsize=(10,5))
pos = nx.spring_layout(ego_net)
nx.draw(ego_net, pos, with_labels=True, node_size=10, node_color='plum', edge_color='gray')
plt.title("Ego Network of 'jquery'", fontsize=20, fontweight='bold')
plt.plot()
```

```python
#Identify number of components
components = list(nx.connected_components(G))

print('Number of Components= ', len(components))
components
```

```python
SG = [G.subgraph(comp) for comp in nx.connected_components(G)]
```

```python
#Retrieve number of nodes, density, connectivity for each subgraph
i = 0
for subgraph in SG:
    nodes = list(subgraph.nodes())
    num_nodes = subgraph.number_of_nodes()
    density = round(nx.density(subgraph),2)
    connectivity = nx.is_connected(subgraph)
    print("Subgraph", i, ": nodes =", num_nodes,", density =" ,density, ", connected = ",connectivity)
    i += 1
    
#Subgraph 0 has 102 nodes (the most out of all other subgraphs)
```

```python
#Subgraph 0 with 102 nodes

sub_graph = SG[0]

plt.figure(figsize=(15,8))
pos = nx.spring_layout(sub_graph)
nx.draw(sub_graph, pos, with_labels=True, node_size=10, node_color='blue', edge_color='gray')
plt.title("Components with 102 nodes", fontsize=20, fontweight='bold')
plt.plot
```

```python
#Subgraph 0 Connectivity Measures

subgraph = SG[0]

print("Is the graph connected?", nx.is_connected(sub_graph))

print("Vertex Connectivity:", nx.node_connectivity(sub_graph))
print("Edge Connectivity:", nx.edge_connectivity(sub_graph))

print("Minimum Nodge Cut:", nx.minimum_node_cut(sub_graph))
print("Minimum Edge Cut:", nx.minimum_edge_cut(sub_graph))
```

# Heatmap Visual of the Network Graphs

```python
def draw(G, pos, measures, measure_name):
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250, cmap=plt.cm.plasma, 
                                   node_color=list(measures.values()),
                                   nodelist=measures.keys())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1, base=10))
    
    # labels = nx.draw_networkx_labels(G, pos)
    edges = nx.draw_networkx_edges(G, pos)

    plt.title(measure_name)
    plt.colorbar(nodes)
    plt.show()
```

```python
draw(G, pos,nx.degree_centrality(G),'Degree Centrality')
```
